Association Rule Mining adalah teknik untuk mengidentifikasi hubungan yang mendasari antara objek yang berbeda. Contoh Supermarket di mana pelanggan dapat membeli berbagai produk. Biasanya ada pola untuk apa yang dibeli pelanggan. Misalnya, ibu dengan bayi membeli produk bayi seperti susu dan popok. Anak perempuan dapat membeli produk rias, sementara bapak yang membeli bir dan keripik, dll. Singkatnya, transaksi melibatkan sebuah pola. Dimungkinkan untuk mendapatkan lebih banyak keuntungan  jika hubungan antara produk yang dibeli dari toko yang berbeda dapat diidentifikasi.

Ada beberapa hal yang perlu diperhatikan untuk sebuah aturan Association Rule Mining, antara lain: Kedua parameter tersebut adalah support dan confidence. Nilai support atau support adalah persentase dari kombinasi item-item tersebut dalam database, sedangkan nilai confidence atau assurance adalah kekuatan hubungan antar item dalam aturan asosiatif.

Beberapa konsep tentang aturan Association Rule Mining yaitu:

1. A=>B

Kami menyebut "A" sebagai "LHS (sisi kiri)," dan "B" sebagai "RHS (sisi kanan)"
Mari kita asumsikan bahwa A adalah popok dan B adalah bir. Artinya ketika seorang pelanggan membeli popok, dia akan membeli bir juga.

2. Support


Dalam hal ini, dukungan berarti kemungkinan pelanggan membeli popok dan bir bersama-sama di antara semua transaksi penjualan.

3. Confidence


Misalkan jika seorang pelanggan mengambil popok. Bagaimana kemungkinan dia untuk membeli bir? Jawabannya adalah “keyakinan” Nilai maksimal dari keyakinan harus 1.

4. Lift


Lift adalah perbandingan sebenarnya antara model naif dan model kami, artinya seberapa besar kemungkinan pelanggan membeli keduanya, dibandingkan membeli secara terpisah, Lift 1 berarti, pelanggan kami cenderung membeli popok dan bir bersama-sama seperti membelinya secara terpisah. Umumnya, agar bermakna dalam pemasaran, lift harus lebih besar dari 1.

Ada 4 tipe dalam association rules pada data mining. Yaitu:
- Multi-relational association rules
- Generalized Association rule
- Interval Information Association Rules
- Quantitative Association Rules

### Multi-Relational Association Rule

Juga dikenal sebagai MRAR, Multi-Relational Association Rule didefinisikan sebagai kelas baru aturan asosiasi yang biasanya berasal dari database yang berbeda atau multi-relasional. Setiap aturan di bawah kelas ini memiliki satu entitas dengan hubungan berbeda yang mewakili hubungan tidak langsung antar entitas. 

### Generalized Association Rule

sebagian besar digunakan untuk mendapatkan gambaran kasar tentang pola menarik yang sering cenderung tetap tersembunyi dalam data. 

### Quantitative Association Rules

Jenis khusus ini sebenarnya adalah salah satu jenis paling unik dari keempat aturan asosiasi yang tersedia. Apa yang membedakannya dari yang lain adalah adanya atribut numerik dalam setidaknya satu atribut aturan asosiasi kuantitatif. Ini berbeda dengan aturan asosiasi umum, di mana sisi kiri dan kanan terdiri dari atribut kategoris. 

Algoritma yang dapat digunakan dalam association rules data mining. Yaitu:
- Apriori Algorithm
- Eclat Algorithm
- FP-Growth Algorithm

### Contoh penggunaan Eclat Algorithm

```python
#Install library ECLAT
!pip install pyECLAT
```

```python
# store the item sets as lists of strings in a list
transactions = [
    ['beer', 'wine', 'cheese'],
    ['beer', 'potato chips'],
    ['eggs', 'flower', 'butter', 'cheese'],
    ['eggs', 'flower', 'butter', 'beer', 'potato chips'],
    ['wine', 'cheese'],
    ['potato chips'],
    ['eggs', 'flower', 'butter', 'wine', 'cheese'],
    ['eggs', 'flower', 'butter', 'beer', 'potato chips'],
    ['wine', 'beer'],
    ['beer', 'potato chips'],
    ['butter', 'eggs'],
    ['beer', 'potato chips'],
    ['flower', 'eggs'],
    ['beer', 'potato chips'],
    ['eggs', 'flower', 'butter', 'wine', 'cheese'],
    ['beer', 'wine', 'potato chips', 'cheese'],
    ['wine', 'cheese'],
    ['beer', 'potato chips'],
    ['wine', 'cheese'],
    ['beer', 'potato chips']
]
```

```python
#import library
import pandas as pd
from pyECLAT import ECLAT

# you simply convert the transaction list into a dataframe
data = pd.DataFrame(transactions)
data
```

```python
# we are looking for itemSETS
# we do not want to have any individual products returned
min_n_products = 2

# we want to set min support to 7
# but we have to express it as a percentage
min_support = 7/len(transactions)

# we have no limit on the size of association rules
# so we set it to the longest transaction
max_length = max([len(x) for x in transactions])
```

```python
# create an instance of eclat
my_eclat = ECLAT(data=data, verbose=True)

# fit the algorithm
rule_indices, rule_supports = my_eclat.fit(min_support=min_support,
                                           min_combination=min_n_products,
                                           max_combination=max_length)
```

```python
print(rule_supports)
```

### Contoh penggunaan FP Growth  Algorithm

```python
# to install the package
!pip install mlxtend

# just in case you are running on Google Colab, you may run into a problem later on if you do not upgrade the package
%pip install mlxtend --upgrade
```

```python
# store the item sets as lists of strings in a list
transactions = [
    ["beer", "wine", "cheese"],
    ["beer", "potato chips"],
    ["eggs", "flower", "butter", "cheese"],
    ["eggs", "flower", "butter", "beer", "potato chips"],
    ["wine", "cheese"],
    ["potato chips"],
    ["eggs", "flower", "butter", "wine", "cheese"],
    ["eggs", "flower", "butter", "beer", "potato chips"],
    ["wine", "beer"],
    ["beer", "potato chips"],
    ["butter", "eggs"],
    ["beer", "potato chips"],
    ["flower", "eggs"],
    ["beer", "potato chips"],
    ["eggs", "flower", "butter", "wine", "cheese"],
    ["beer", "wine", "potato chips", "cheese"],
    ["wine", "cheese"],
    ["beer", "potato chips"],
    ["wine", "cheese"],
    ["beer", "potato chips"],
]
```

```python
# it is necessary for mlxtend to reorganise the data
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

# instantiate a transaction encoder
my_transactionencoder = TransactionEncoder()

# fit the transaction encoder using the list of transaction tuples
my_transactionencoder.fit(transactions)

# transform the list of transaction tuples into an array of encoded transactions
encoded_transactions = my_transactionencoder.transform(transactions)

# convert the array of encoded transactions into a dataframe
encoded_transactions_df = pd.DataFrame(encoded_transactions, columns=my_transactionencoder.columns_)
encoded_transactions_df
```

```python
# our min support is 7, but it has to be expressed as a percentage for mlxtend
min_support = 7/len(transactions) 

# compute the frequent itemsets using fpgriowth from mlxtend
from mlxtend.frequent_patterns.fpgrowth import fpgrowth
frequent_itemsets = fpgrowth(encoded_transactions_df, min_support=min_support, use_colnames = True)

# print the frequent itemsets
frequent_itemsets
```

```python
# Compute the association rules based on the frequent itemsets
from mlxtend.frequent_patterns import association_rules

# compute and print the association rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

#### Kesimpulan
Kita dapat menyimpulkan bahwa ada dua kombinasi produk, dan kedua asosiasi adalah dua arah. Orang yang membeli keju, juga membeli anggur dan orang yang membeli anggur juga membeli keju. Secara terpisah, kita melihat bahwa orang yang membeli bir juga membeli keripik kentang dan sebaliknya.
